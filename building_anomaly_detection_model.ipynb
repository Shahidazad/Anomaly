```python
# from google.colab import drive
import os, sys
# drive.mount('/content/drive')
import numpy as np
import pandas as pd
import time
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
```

```python
# os.chdir()
data=pd.read_csv('unsupervised_deep_learning_dataset.csv')
```

```python
for i in data.columns:
    if(data[i].dtype == 'object'):
        data[i]=LabelEncoder().fit_transform(data[i])
    
```

```python
list1=[]
for i in data.columns:
    null  = sum(pd.isnull(data[i]))
    null1=i+' - '+str(null)
    list1.append(null1)
list1
```

```python
mean=np.mean(data,axis=0)
std=np.std(data,axis=0)
```

```python
data=data - mean
```

```python
data =data /std
```

```python
data
```

```python
input_layer =tf.keras.layers.Input(shape=(data.shape[1],))
encoded=tf.keras.layers.Dense(128,activation="relu")(input_layer)
#encoding layer with LSTM units and ReLU activation function
encoded=tf.keras.layers.Dense(64,activation="relu")(encoded)
decoded=tf.keras.layers.Dense(128,activation="relu")(encoded)
decoded=tf.keras.layers.Dense(data.shape[1],activation="sigmoid")(decoded)
```

```python
autoencoder=tf.keras.models.Model(input_layer,decoded)
```

```python
autoencoder.compile(optimizer="adam",loss="mean_squared_error")
```

```python
autoencoder.fit(data,data,epochs=150,batch_size=65)
```

```python
# CALCULATE RECONSTRUCTION ERROR
predict=autoencoder.predict(data)
recon_error =np.mean((predict-data)**2,axis=1)
```

```python
# thereshold calculation 
threshold=np.mean(recon_error) +3*np.std(recon_error)
```

```python
anomalies=np.where(recon_error > threshold)
```

```python
data.iloc[anomalies]*std        + mean
```

```python
data*std        + mean
```
